Case study 1: Student Performance Analysis

```python
data = {
    "Id" : [1,2,3,4,5],
    "Name" : ["A","B","C","D","E"],
    "Marks" : [55,78,92,56,78]
}

df = pd.DataFrame(data)
df

#Display> 75 Marks
print(df[df["Marks"] > 75])

#Add grade to the students record
df["Grade"] = df["Marks"].apply(lambda x: "A" if x > 75 else "B")
print(df)

#Lambda: topper
#Filter the data for 'A' grade
#Add NCC marks 3

```
